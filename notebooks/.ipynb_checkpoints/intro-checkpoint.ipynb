# Проект KitKat


О чём это? <br>
*Совсем кратко*: парсер данных.

*Подробнее*: kitkat собирает у **поставщиков**<sup><font size='small'>1)<font></sup>&nbsp; актуальные данные о позициях (**товарах**<sup><font size='small'>2)<font></sup>&nbsp;) и обновляет каталог **клиента**<sup><font size='small'>3)<font></sup>&nbsp;. 

<sup><font size='small'>1)<font></sup> ***Поставщик*** - Любой источник информации о **товаре**. Поставщиком может быть таблица (файл Excel, csv, Google таблицы),<br>&ensp;API или парсинг страниц HTML. В коде он определен классом **Supplier**. Supplier запускет сценарий, собирает ответ и отдает cполученные данные сущности **товер**

<sup><font size='small'>2)<font></sup> ***Товар*** - Полученные запрошенные данные. В коде он определен классом **Product**.<br>
&ensp;Товаром может быть материальный объект, биржевая, спортивная или научная информация. kitkat умеет мониторить изменения параметров товара.<br> &ensp;Товар (*Product*) получает сырые данные от поставщика (*Supplier*), форматирует их и отдает опеделенному **клиенту**.

<sup><font size='small'>3)<font></sup> ***Клиент*** - получатель информации о тевре. В каестве клиента может выступать база данных, <br>
&ensp;таблица (файл Excel, csv, Google таблицы) или email.


*Пример автоматизации с помощью kitkat*: <br>
Агент по закупкам получает запрос от исследовательского отдела найти на рынке компонент Y, с параметрами a,b,c. Типичный алгоритм действия:

- Первичный поиск через поисковые движки (Googlе, Bing)
- Поиск в каталогах (mouser, digikey)
- Отбор позиций по параметрам
- Анализ полученных данных
